Using your friends initial research, you'll delve into the Netflix data to see if you can determine whether movie lengths are actually getting shorter and explain some of the contributing factors, if any.

```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Start coding!
```

```python
# 01 Load and inspect the Netflix data

# Create a dataframe from a csv
netflix_df = pd.read_csv("netflix_data.csv")

# View the first five rows
netflix_df.head(5)
```

```python
# 02 Subset the DataFrame for "Movies"

# Subset only rows where the "type" is a "Movie"
netflix_subset = netflix_df[netflix_df["type"] == "Movie"]

# Print
netflix_subset.head()
```

```python
# 03 Subset the columns of the new dataframe

netflix_movies = netflix_subset[["title", "country", "genre",
                                "release_year", "duration"]]

netflix_movies.head()
```

```python
# 04 Filter the dataframe by movie duration

# movies with duration of fewer than 60 mins
short_movies = netflix_movies[netflix_movies["duration"] < 60]

# print first 20 rows
short_movies.head(20)
```

```python
# 05 Assign colors to movie genres

# add colors to list "colors" based on genre

colors = []

for movie, row in netflix_movies.iterrows():
    if row["genre"] == "Children":
        colors.append("white")
    elif row["genre"] == "Documentaries":
        colors.append("red")
    elif row["genre"] == "Stand-Up":
        colors.append("black")
    else:
        colors.append("blue")
        
print(colors[:20])
```

```python
# 06 Plot the movie duration by year using the colors created

# creating a figure
fig = plt.figure(figsize=(12, 8))

# creating scatter plot
plt.scatter(netflix_movies["release_year"], netflix_movies["duration"], c = colors)

# add labels
plt.xlabel("Release year")
plt.ylabel("Duration (min)")

# add title
plt.title("Movie Duration by Year of Release")
```
